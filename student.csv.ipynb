```python
# grading the student
df['average score'] = (df['math score'] + df['reading score'] +df['writing score'])/3
round(df['average score'].mean(),2)
```

```python
df
```

```python
def rating(score):
    if score >=90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    elif score >= 40:
        return "E"
    else:
        return "F"
df['rating'] = df['average score'].apply(rating)
df


```
