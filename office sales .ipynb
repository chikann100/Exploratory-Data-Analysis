```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dm_office_sales.csv')
df
```

```python
sns.scatterplot(data = df, x = 'sales', y = 'salary', hue ='division')
#  output : this shows that there is a correlation between sales and salary
```

```python
sns.histplot(df, x = 'salary')
# output: this shows a division in salary
```

```python
plt.figure(figsize =(10,5))
sns.countplot(df, x ='division')
```

```python
plt.figure(figsize =(10,5))
sns.barplot(df, x = 'division', y ='salary')
```

```python
plt.figure(figsize = (10,5))
sns.boxplot(x = 'division', y = 'salary', data = df)
```
