```python
# average mean number of nights for a stay across the entire data

df['total_stay'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']
round(df['total_stay'].mean(),2)
```

```python
#what is the average total cost for a stay in the dataset
df['total_cost'] = df['adr'] * df['total_stay']
round(df['total_cost'].mean(),2)
```

```python
def arrival(x):
    count = 0
    for i in range(1,16):
        for j in x:
            if j == i:
                count += 1
    return count
arrival([10, 2, 13, 30, 2])
```

```python
arrival(df['arrival_date_day_of_month'])
```
